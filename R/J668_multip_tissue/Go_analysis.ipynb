```r
## 分析lemon-tree 聚类结果的Go term
rm=list(ls())
setwd('G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/cluster')
library(clusterProfiler)
library(org.At.tair.db)
df <- read.table("lemon-tree_clusters.txt", sep='\t', header=T)
for (i in seq(0,330))
{
    df_cluster = df[df$cluster_id==i, ]
    output_f = paste("lemon-tree_cluster_", as.character(i), ".txt", sep="")
    df_cluster_gid = df_cluster$Atha_id
    ego = enrichGO(df_cluster_gid, OrgDb = "org.At.tair.db", keyType="TAIR", ont = "ALL", pAdjustMethod = "BH", pvalueCutoff  = 0.01, qvalueCutoff  = 0.05, readable= TRUE)
    write.table(summary(ego), output_f, sep='\t', row.names=FALSE)
}
```

```r
## 对Go term 结果作图
rm(list = ls())
options(repr.plot.width=15,repr.plot.height=6)
setwd('G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/cluster/lemon_cluster_Go/A_go_figure')
library(ggplot2)
library(ggpubr)
df <- read.csv('lemon-tree_cluster_182_go.txt', sep = '\t', header = T)
fg <- ggplot(df, aes(x=-log10(FDR), y=bioprocess, size=gene_number, color=-log10(FDR)))+
geom_point()+
labs(x=NULL, y=NULL)+
theme(panel.background = element_blank(), 
      panel.grid = element_blank(),
      axis.line=element_line(linetype=1,color="black"),
      #legend.position=c(0.25, 0.7),
      axis.ticks.length = unit(0.5, 'cm'),
      axis.text = element_text(size = 16)
     )+
scale_color_gradient(low="#F0AB97",  high="red")+
scale_size_area(max_size = 10)
fg
ggsave(plot =fg, filename = 'lemon-tree_cluster_182_go.pdf', height =8, width = 12.5)
```
