```r
setwd("G:/Billfish/J668_multip_tissue_3D-genome/HiC/TAD/TAD_size")
library(tidyverse)
library(ComplexHeatmap)
library(circlize)
library(RColorBrewer)



F1 <- read.table("all_tissues_TAD_size.bed",header = T)

t <- "Anther"
output_f = paste(t, "_At.pdf", sep="")
F2 <- F1[F1$Subgenome=='At' & F1$Tissue==t, ]
#print(F2)
df <- data.frame(table(cut(F2$Size,
                          breaks = c(seq(0,1000,by = 20))))) %>%
  setNames(c("seq","Freq"))
df <- df %>%
  mutate(ratio = Freq / sum(df$Freq)) %>%
  select(1,3) %>%
  mutate(seq = factor(seq)) %>%
  column_to_rownames("seq") 
col_fun = colorRamp2(seq(0,0.1,0.0001), colorRampPalette(brewer.pal(9, "PuRd"))(1001))
df <- t(df)
#print(df)
pdf(output_f)
Heatmap(df, 
            #column_names_side = "top",
            col = col_fun,
            show_row_names = FALSE,
            show_column_names = FALSE,
            cluster_rows = FALSE,
            cluster_columns = FALSE,
            use_raster = TRUE ,
            border = TRUE,
            heatmap_legend_param = list(legend_height=unit(8,"cm"),
                                        legend_direction="horizontal",
                                        leged_position = "bottom"))
dev.off()


```
