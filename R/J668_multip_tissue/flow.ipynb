```r
setwd("G:/Billfish/J668_multip_tissue_3D-genome/HiC/loops/loop_conserved_dynamic")
flower_plot <- function(sample, value, start, a, b, 
	ellipse_col = rgb(135, 206, 235, 150, max = 255), 
	circle_col = rgb(0, 162, 214, max = 255),
	circle_text_cex = 1.5
	) {
	par( bty = "n", ann = F, xaxt = "n", yaxt = "n", mar = c(1,1,1,1))
	plot(c(0,10),c(0,10),type="n")
	n   <- length(sample)
	deg <- 360 / n

    color_vaule = c("#1FA675", "#D95204", "#7571B2", "#E6298D", "#A6751B", "#69A71E", "#E2AE09", "#E6191E", "#1E79B2", "#df6b24", "#419FB7", "#efdab4")

	res <- lapply(1:n, function(t){
		draw.ellipse(x = 5 + cos((start + deg * (t - 1)) * pi / 180), 
			         y = 5 + sin((start + deg * (t - 1)) * pi / 180), 
			         col = color_vaule[t],
			         border =circle_col,
			         a = a, b = b, angle = deg * (t - 1))
		# text(x = 5 + 2.5 * cos((start + deg * (t - 1)) * pi / 180),
		# 	 y = 5 + 2.5 * sin((start + deg * (t - 1)) * pi / 180),
		# 	 value[t]
		# 	)
        if (deg * (t - 1) < 180 && deg * (t - 1) > 0 ) {
                    text(x = 5 + 1.3 * cos((start + deg * (t - 1)) * pi / 180),
                        y = 5 + 1.3 * sin((start + deg * (t - 1)) * pi / 180),
                        value[t],
                        srt = deg * (t - 1) - start,
                        adj = 1,
                        cex = circle_text_cex
                        )

                } else {
                    text(x = 5 + 1.3 * cos((start + deg * (t - 1)) * pi / 180),
                        y = 5 + 1.3 * sin((start + deg * (t - 1)) * pi / 180),
                        value[t],
                        srt = deg * (t - 1) + start,
                        adj = 0,
                        cex = circle_text_cex
                        )
                }			


		if (deg * (t - 1) < 180 && deg * (t - 1) > 0 ) {
			text(x = 5 + 3.3 * cos((start + deg * (t - 1)) * pi / 180),
				 y = 5 + 3.3 * sin((start + deg * (t - 1)) * pi / 180),
				 sample[t],
				 srt = deg * (t - 1) - start,
				 adj = 1,
				 cex = circle_text_cex
				)

		} else {
			text(x = 5 + 3.3 * cos((start + deg * (t - 1)) * pi / 180),
				 y = 5 + 3.3 * sin((start + deg * (t - 1)) * pi / 180),
				 sample[t],
				 srt = deg * (t - 1) + start,
				 adj = 0,
				 cex = circle_text_cex
				)
		}			
	})
	draw.circle(x = 5, y = 5, r = 1.3, col = circle_col, border = circle_col)
}
```
