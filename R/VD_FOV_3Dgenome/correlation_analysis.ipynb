```r
rm(list = ls())
setwd('F:/VD_FOV_3Dgenome/RNA_Seq/Rep')
library(Hmisc)
library(corrplot)
library(pheatmap)
data <- read.csv('glr19_VD_FOV_TPM.txt', sep = '\t', header = T)
data_t <- data[,2:31]
res <- rcorr(as.matrix(data_t))
CorMatrix <- function(cor, p) {
  ut <- upper.tri(cor)
  data.frame(row = rownames(cor)[row(cor)[ut]],
             column = rownames(cor)[col(cor)[ut]],
             cor = (cor)[ut],
             p = p[ut])
}
RNA_CorMatrix <- CorMatrix(res$r, res$P)
options(repr.plot.width=8, repr.plot.height=8)
p <- corrplot(res$r,  col=colorRampPalette(c("#00AFA6","#FFE7B3","#ee3f4d"))(200), is.corr=FALSE, tl.col='black')
```
