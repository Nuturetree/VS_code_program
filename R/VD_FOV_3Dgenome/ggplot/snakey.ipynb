```r
## 基因差异表达动态变化桑基图
rm(list=ls())
library(ggplot2)
library(alluvial)
setwd('F:/VD_FOV_3Dgenome/RNA_Seq/DEG/J668_mock_Fov7')
tit = read.csv('J668_mock_Fov7_snaky_format.txt', header = T, sep = '\t')
alluvial(tit[,1:7], freq=tit$number,
         #col = tit$DPA20,
         #col = ['#72D9C6','#C3D9B8','#F1C848'],
         #border = tit$DPA20,
         col = ifelse(tit$h12 == "Down", '#FFE9AA', ifelse(tit$h12 == "Up", "#D94854", 'grey')),
         border = ifelse(tit$h12 == "Down", '#FFE9AA', ifelse(tit$h12 == "Up", "#D94854", 'grey')),
         #border = ifelse(tit$DPA20 == "Yes", "orange", "grey"),
         hide = tit$number == 0,
         cex = 0.7
)
```

```r
pdf(file = 'J668_mock_Fov7_snaky_format.pdf')
alluvial(tit[,1:7], freq=tit$number,
         #col = tit$DPA20,
         #col = ['#72D9C6','#C3D9B8','#F1C848'],
         #border = tit$DPA20,
         col = ifelse(tit$h12 == "Down", '#FFE9AA', ifelse(tit$h12 == "Up", "#D94854", 'grey')),
         border = ifelse(tit$h12 == "Down", '#FFE9AA', ifelse(tit$h12 == "Up", "#D94854", 'grey')),
         #border = ifelse(tit$DPA20 == "Yes", "orange", "grey"),
         hide = tit$number == 0,
         cex = 0.7
)
dev.off()
```
