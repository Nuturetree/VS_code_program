```r
## 
rm(list=ls())
library(ggplot2)
library(alluvial)

setwd('F:/TM-1_leaf_3D-genome/Proe-C_HiC/TAD/Cliques/gene')
tit = read.csv('PoreC_HiC_gene_dynamic.txt', header = T, sep = '\t')
tit <- within(tit, hic <- factor(hic, levels = c("nocliques", "3.0", "4.0", "5.0", '6.0', '7.0')))
with(tit, levels(hic))
tit <- within(tit, porec <- factor(porec, levels = c("nocliques", "3.0", "4.0", "5.0", '6.0', '7.0','8.0', '9.0', '10.0', '11.0', '12.0', '13.0', '14.0', '15.0', '16.0', '17.0', '18.0', '19.0', '20.0', '21.0', '22.0', '>22')))
with(tit, levels(porec))
#print(tit)
alluvial(tit[,1:2], freq=tit$Geneid,
         #col = tit$DPA20,
         #col = ['#72D9C6','#C3D9B8','#F1C848'],
         #border = tit$DPA20,
         col = ifelse(tit$hic == "3.0", '#D4BCB0', ifelse(tit$hic == "4.0",  '#C42B2E', ifelse(tit$hic == '5.0','#F60000', ifelse(tit$hic == '6.0', '#4399A9', ifelse(tit$hic == 'nocliques', 'grey', '#751A1B'))))),
         border = ifelse(tit$hic == "3.0", '#D4BCB0', ifelse(tit$hic == "4.0",  '#C42B2E', ifelse(tit$hic == '5.0','#F60000', ifelse(tit$hic == '6.0', '#4399A9', ifelse(tit$hic == 'nocliques', 'grey', '#751A1B'))))),
         #border = ifelse(tit$DPA20 == "Yes", "orange", "grey"),
         hide = tit$Geneid == 0,
         cex = 0.7
)
```
