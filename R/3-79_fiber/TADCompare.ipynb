```r
## 分析比较四个时期TAD的保守性 , 0 vs 10
library(dplyr)
library(SpectralTAD)
library(TADCompare)
chrs = c('Gbar_A01', 'Gbar_A02', 'Gbar_A03', 'Gbar_A04', 'Gbar_A05', 'Gbar_A06', 'Gbar_A07', 'Gbar_A08', 'Gbar_A09', 'Gbar_A10', 'Gbar_A11', 'Gbar_A12', 'Gbar_A13',
         'Gbar_D01', 'Gbar_D02', 'Gbar_D03', 'Gbar_D04', 'Gbar_D05', 'Gbar_D06', 'Gbar_D07', 'Gbar_D08', 'Gbar_D09', 'Gbar_D10', 'Gbar_D11', 'Gbar_D12', 'Gbar_D13')
for (c in chrs) {
    mtx_f1 = paste("F:/fiber/airticle/GB_modification/TADCompare/0DPA/",c, '.matrix', sep = "")
    abs_f1 = paste("F:/fiber/airticle/GB_modification/TADCompare/0DPA/",c, '_abs.bed', sep = "")
    mtx_f2 = paste("F:/fiber/airticle/GB_modification/TADCompare/10DPA/",c, '.matrix', sep = "")
    abs_f2 = paste("F:/fiber/airticle/GB_modification/TADCompare/10DPA/",c, '_abs.bed', sep = "")
    out_f = paste("F:/fiber/airticle/GB_modification/TADCompare/0DPA_10DPA/",c, '_TADcompare.bed', sep = "")
    mat1 = read.table(mtx_f1)
    bed1 = read.table(abs_f1)
    mat2 = read.table(mtx_f2)
    bed2 = read.table(abs_f2)
    #View(bed1)
    sparse_mats1 <- HiCcompare::hicpro2bedpe(mat1, bed1)
    sparse_mats2 <- HiCcompare::hicpro2bedpe(mat2, bed2)
    x <- sparse_mats1$cis[[1]]
    y <- sparse_mats2$cis[[1]]
    x <- x[, c(2, 5, 7)]
    y <- y[, c(2, 5, 7)]
    comp <- TADCompare(x, y, resolution = 20000)
    df_out <- comp$TAD_Frame
    write.table(df_out, out_f, sep = '\t', row.names = FALSE)
}
```
