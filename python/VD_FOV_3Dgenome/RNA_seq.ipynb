```python
## 寻找稳定的差异表达基因动态变化过程
df_out = df_six.groupby(by=['min1', 'min5', 'min10', 'min30', 'min90', 'min180', 'h12'], as_index=False)['GeneId'].count()

df_out['tags']= df_out['min1'] + df_out['min5'] + df_out['min10'] +df_out['min30'] +df_out['min90'] + df_out['h12']
print(df_out.shape)
import re
patern = re.compile('Not')
df_result = pd.DataFrame(columns=['min1', 'min5', 'min10', 'min30', 'min90', 'min180', 'h12', 'number'])
i=0
for index, row in df_out.iterrows():
    t = row['tags']
    l = len(patern.findall(t))
    if l<4:
        df_result.loc[i, :] = [row['min1'], row['min5'], row['min10'], row['min30'], row['min90'], row['min180'], row['h12'], row['GeneId']]
        i+=1
print(df_result.shape)
#df_result.to_csv('F:/VD_FOV_3Dgenome/RNA_Seq/DEG/J668_mock_Fov7/J668_mock_Fov7_snaky_format_not.txt', sep='\t', index=False)
#df_result.to_csv('F:/VD_FOV_3Dgenome/RNA_Seq/DEG/J668_mock_v991/J668_mock_v991_snaky_format_not.txt', sep='\t', index=False)
df_result.to_csv('F:/VD_FOV_3Dgenome/RNA_Seq/DEG/glr19_mock_Fov7/glr19_mock_Fov7_snaky_format_not.txt', sep='\t', index=False)
```
