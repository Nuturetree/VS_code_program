```python
## 分析时期特异性高表达基因 (TAU)
import pandas as pd
import numpy as np
# 1、数据预处理
def TAU_data_format(df):
    times = ['0min', '1min', '5min', '10min', '30min', '90min', '180min', '12h']
    #times = ['0min', '30min',  '90min', '180min', '12h']
    tags = ['GeneId', 'tags']
    df_tags = df[tags].copy()
    df_times = df[times].copy()
    df_times[df_times<=1] = 1
    for t in times:
        df_times[t] = df_times[t].apply(np.log)
    df_return = pd.concat([df_tags, df_times], axis=1)
    df_return.replace(0, 0.0001, inplace=True)
    return df_return

# 2、计算TAU
def TAU_calculate(df):
    #print(df)
    for index, row in df.iterrows():
        n = 8 ## The number of time
        #print(row.tolist())
        #print(row)
        tissue_tpm = sorted(row[2:10])
        #print(tissue_tpm)
        tau = n/(n-1) - sum(tissue_tpm)/((n-1)*max(tissue_tpm)) # Calculate the tau
        df.loc[index, 'TAU'] = tau
    return df
df = pd.read_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/Rep/J668_j668_VD_Fov7_averge_TPM_long.txt", sep="\t")
```

```python
df_0_30_180_12h = df[['GeneId', 'tags', '0min', '1min', '5min', '10min', '30min', '90min', '180min', '12h']]
df_TAU_data = TAU_data_format(df_0_30_180_12h) # 计算TAU 前数据预处理
df_TAU = TAU_calculate(df_TAU_data) # 计算TAU 
```

```python
import pandas as pd
# 数据z-score 标准化
def TPM_zscore(df):
    times = ['0min', '1min', '5min', '10min', '30min', '90min', '180min', '12h']
    tags = ['GeneId', 'tags']
    df_tags = df[tags].copy()
    df_times = df[times].copy()
    df_tmp = pd.DataFrame()
    for i in range(df_times.shape[0]):
        df_normal = df_times.iloc[[i]].apply(lambda x: (x-np.mean(x))/(np.std(x)), axis=1)
        df_tmp = df_tmp.append(df_normal)
    df_return = pd.concat([df_tags, df_tmp], axis=1) 
    return df_return
df_TPM = pd.read_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/Rep/J668_j668_VD_Fov7_averge_TPM_long.txt", sep="\t")
df_TAU = pd.read_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/TAU/J668_J668_TAU.bed", sep="\t")
df_TAU_08 = df_TAU.query("TAU>=0.8")[['GeneId', 'tags']].copy() # 选择时期特异表达基因
df_TPM_TAU_08 = pd.merge(df_TPM, df_TAU_08, how='inner', on=['GeneId', 'tags'])[['GeneId', 'tags', '0min', '1min', '5min', '10min', '30min', '90min', '180min', '12h']].copy()
df_zscore = TPM_zscore(df_TPM_TAU_08)
df_zscore.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/TAU/J668_glr_0.8_TAU.bed", sep="\t", index=False)
```

```python
## 分析有多少时期特异性基因，在不同时期具有多少基因具有表达差异
import pandas as pd
df = pd.read_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/TAU/J668_glr_TAU_umap_zscore.bed", sep="\t")
df_glr_mock = df.query("tags=='glr_mock'").copy()
df_glr_fov7 = df.query("tags=='glr_fov7'").copy()
df_j668_mock = df.query("tags=='j668_mock'").copy()
df_j668_fov7 = df.query("tags=='j668_fov7'").copy()
df_j668_v991 = df.query("tags=='j668_v991'").copy()
def up_down(df_one, df_two, tags1, tags2):
    times = ['0min', '1min', '5min', '10min', '30min', '90min', '180min', '12h']
    df_return1 = pd.DataFrame(columns=['tags1', 'tag2', 'time', 'down', 'up'])
    df_return2 = pd.DataFrame(columns=[])
    for t in times:
        df_time1, df_time2 = df_one.query("max_tpm==@t")[['GeneId', 'tags']], df_two.query("max_tpm==@t")[['GeneId', 'tags']]
        df_tmp = pd.concat([df_time1, df_time2])
        df_tmp.drop_duplicates(keep=False, subset=['GeneId'], inplace=True)
        df_down = df_tmp.query("tags==@tags1").copy()
        df_up = df_tmp.query("tags==@tags2").copy()
        df_return1.loc[df_return1.shape[0]] = [tags1, tags2, t, df_down.shape[0], df_up.shape[0]]
        ## 将所有时期特有基因选出
        df_down['sig']   = ['down']*df_down.shape[0]
        df_down['times'] = [t]*df_down.shape[0]
        df_up['sig']   = ['up']*df_up.shape[0]
        df_up['times'] = [t]*df_up.shape[0]
        df_return2 = pd.concat([df_return2, df_down, df_up])
        df_return2['tags']  = ["{0}_{1}".format(tags1, tags2)]*df_return2.shape[0]
        #print(df_return2)
    return df_return1, df_return2
glr_mock_vs_fov7, glr_mock_vs_fov7_2 = up_down(df_glr_mock, df_glr_fov7, 'glr_mock', 'glr_fov7')
j668_mock_vs_fov7, j668_mock_vs_fov7_2 = up_down(df_j668_mock, df_j668_fov7, 'j668_mock', 'j668_fov7')
j668_mock_vs_v991, j668_mock_vs_v991_2 = up_down(df_j668_mock, df_j668_v991, 'j668_mock', 'j668_v991')
mock_glr_vs_j668, mock_glr_vs_j668_2 = up_down(df_glr_mock, df_j668_mock, 'glr_mock', 'j668_mock')
fov7_glr_vs_j668, fov7_glr_vs_j668_2 = up_down(df_glr_fov7, df_j668_fov7, 'glr_fov7', 'j668_fov7')
df_result = pd.concat([glr_mock_vs_fov7, j668_mock_vs_fov7, j668_mock_vs_v991, mock_glr_vs_j668, fov7_glr_vs_j668])
df_result.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/TAU/all_time_diff_gene_number.bed", sep="\t", index=False)

df_result2 = pd.concat([glr_mock_vs_fov7_2, j668_mock_vs_fov7_2, j668_mock_vs_v991_2, mock_glr_vs_j668_2, fov7_glr_vs_j668_2])
## 替换标签
old_sig = ['glr_mock_glr_fov7', 'j668_mock_j668_fov7', 'j668_mock_j668_v991', 'glr_mock_j668_mock', 'glr_fov7_j668_fov7']
new_sig = ['Glr_mock_fov7', 'J668_mock_fov7', 'J668_mock_v991', 'Mock_Glr_J668', 'Fov7_Glr_J668']
df_result2['tags'].replace(old_sig, new_sig, inplace=True)
df_result2.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/TAU/all_time_diff_gene.bed", sep="\t", index=False)
```

```python
##  分析差异表达基因类型， 同时上调，同时下调，上下调不一致，特有上下调 (差异表达基因分类)
import pandas as pd
def DEG_type(df1, df2, tags1, tags2):
    df = pd.concat([df1,df2], axis=0)
    df_TAIR = pd.read_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/Gh_vs_Arabidopsis.txt", sep="\t", names=['GeneId', 'TAIRID'])
    #! 特有上下调
    df_uniq = df.drop_duplicates(keep=False, subset=['GeneId', 'times']) # 把一个基因id 同时刻出现两次 剔除。
    df_uniq_tags1 = df_uniq.query("tags==@tags1")[['GeneId', 'times']].copy()
    df_uniq_tags2 = df_uniq.query("tags==@tags2")[['GeneId', 'times']].copy()
    print(df_uniq_tags1.shape[0] + df_uniq_tags2.shape[0], df_uniq.shape[0])
    df_uniq_tags1['tags'] = ['BF_uniq']*df_uniq_tags1.shape[0]
    df_uniq_tags2['tags'] = ['LT_uniq']*df_uniq_tags2.shape[0]
    #df_uniq['sig'] = ['Down']
    #! 同时有差异，但上下调方向不一致
    df_all = pd.concat([df, df_uniq], axis=0)
    df_both = df_all.drop_duplicates(keep=False) # 把一个基因id， 同时刻出现两次，保留
    df_both_uniq = df_both.drop_duplicates(keep=False, subset=['GeneId', 'times', 'sig']).copy() # 一个基因Id 同时刻出现两次，并且具有相同 上下调标签，去除
    #df_both_uniq.drop_duplicates(keep='first', subset=['GeneId'], inplace=True)# 保留一个基因，关于上调和下调
    #! 同时有差异，上下调方向相同
    df_both2 = pd.concat([df_both, df_both_uniq], axis=0)
    df_both2.drop_duplicates(keep=False, inplace=True)
    df_both_uniq['tags'] = ['opposite']*df_both_uniq.shape[0]
    df_both_uniq = df_both_uniq[['GeneId' , 'times', 'tags']] 
    df_up = df_both2.query("sig=='Up'").copy()
    df_down = df_both2.query("sig=='Down'").copy()
    print(df.shape[0], df_up.shape[0], df_down.shape[0], df_both_uniq.shape[0], df_uniq_tags1.shape[0], df_uniq_tags2.shape[0])
    print( df_up.shape[0] + df_down.shape[0] + df_both_uniq.shape[0] + df_uniq_tags1.shape[0] + df_uniq_tags2.shape[0])
    df_up['tags'] = ['Up_Up']*df_up.shape[0]
    df_down['tags'] = ['Down_Down']*df_down.shape[0]
    df_up = df_up[['GeneId', 'times', 'tags']]
    df_down = df_down[['GeneId', 'times', 'tags']]
    df_return1 = pd.concat([df_uniq_tags1, df_uniq_tags2, df_both_uniq, df_up, df_down], axis=0) ## 合并所有数据
    df_return2 = df_return1.copy()
    df_return2['number'] = [1]*df_return2.shape[0]
    df_return2.reset_index(drop=True, inplace=True)
    #print(df_return2)
    df_return2 = df_return2.groupby(by=['times', 'tags'], as_index=False)['number'].sum()
    all_time = ['1min', '5min', '10min', '30min', '90min', '180min', '12h']
    df_return3 = pd.DataFrame()
    for t in all_time:
        df_tmp = df_return2.query("times==@t").copy()
        all_number = df_tmp['number'].sum()
        df_tmp['ratio'] = (df_tmp['number']/all_number)*100
        df_return3 = pd.concat([df_return3, df_tmp], axis=0)
    old_time = ['1min', '5min', '10min', '30min', '90min', '180min', '12h']
    new_time = ['1 min', '5 min', '10 min', '30 min', '90 min', '180 min', '720 min']
    df_return1['times'].replace(old_time, new_time, inplace=True)
    df_return3['times'].replace(old_time, new_time, inplace=True)
    ## add TAIRID columns
    df_return1 = pd.merge(df_return1, df_TAIR, how='left', on=['GeneId'])
    return df_return1, df_return3

df_J668_mock_Fov7 = pd.read_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/DEG/J668_mock_Fov7/J668_mock_Fov7_all_min_DEG.bed", sep="\t")
df_J668_mock_Fov7 = pd.read_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/DEG/J668_mock_Fov7/J668_mock_Fov7_all_min_DEG.bed", sep="\t")
df_J668_mock_V991 = pd.read_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/DEG/J668_mock_v991/J668_mock_v991_all_min_DEG.bed", sep="\t")
## 运行，并写出, J668_Fov7_V991
df_return1, df_return2 = DEG_type(df_J668_mock_Fov7, df_J668_mock_V991, "J668_mock_Fov7", "J668_mock_v991")
df_return1.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/DEG/DEG_type/J668_Fov7_V991/J668_Fov7_V991_DEGtype.bed", sep="\t", index=False)
df_return2.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/DEG/DEG_type/J668_Fov7_V991/J668_Fov7_V991_DEGtype_number.bed", sep="\t", index=False)

## Fov7_J668_J668
df_return3, df_return4 = DEG_type(df_J668_mock_Fov7, df_J668_mock_Fov7, 'J668_mock_Fov7', "J668_mock_Fov7")
df_return3.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/DEG/DEG_type/J668_J668_Fov7/J668_J668_Fov7_DEGtype.bed", sep="\t", index=False)
df_return4.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/DEG/DEG_type/J668_J668_Fov7/J668_J668_Fov7_DEGtype_number.bed", sep="\t", index=False)

df_return2['sig'] = ['Fov7_V991 (J668)']*df_return2.shape[0]
df_return4['sig'] = ['J668_J668 (Fov7)']*df_return4.shape[0]
df = pd.concat([df_return2, df_return4], axis=0)
df.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq/DEG/DEG_type/DEG_type_number_J668_Fov7.bed", sep="\t", index=False)
```

```python
## 从Go term 到networkx
#! 1、用GOSemSim 来分析Go term 的相似性 并用simpleEnrichment来聚类
#! 2、用相似性程度作为权重，构造互作网络
#! 3、构建节点的描述信息

#! 2 将矩阵转换为网络格式
def matrix_To_net(df1, df2):
    id_list = df2['id'].tolist()
    df_return = pd.DataFrame(columns=['source', 'target', 'weight'])
    list_len = len(id_list)
    for i in range(list_len):
        for j in range(i+1, list_len ):
            s = id_list[i]
            t = id_list[j]
            w = df1.loc[s, t]
            df_return.loc[df_return.shape[0]] = [s, t, w]
    df_return = df_return.query("weight>=0.1").copy()
    df_return['Type'] = ['Undirected']*df_return.shape[0]
    return df_return

#! 3 节点的描述信息
# df1 Go term 描述信息
# df2 Go term 聚类信息
# df3 用作比较的Go term
def node_descriptions(df1, df2, df3):
    df1 = df1.query("ontology=='biological_process'").copy()
    df_descript = df1[['id', 'description']]
    # 对另外一个处理 构造其方便比较的结构
    df_other_Go = df3[['id']].copy()
    df_other_Go.columns = ['id']
    df_other_Go['YN'] = ['both']*df_other_Go.shape[0]
    # 给节点打上Go描述信息
    df_node = pd.merge(df2, df_descript, how='left', on=['id'])
    # 确定Go terms 是否在两个比较间均有出现
    df_node = pd.merge(df_node, df_other_Go, how='left', on=['id'])
    df_node.fillna('single', inplace=True) # 对空的处理赋值single
    return df_node

## 给出每个簇前三个最具有代表性的 Go_terms
# df1 表示节点描述信息
# df2 表示Go terms 相似矩阵
# N 表示需要对多少个cliques 进行分析
def std_top3(df1, df2, N):
    df_return = pd.DataFrame()
    df_tmp2 = pd.DataFrame()
    for i in range(N):
        j = i+1 # 类是从1 开始
        go_id = df2.query("cluster==@j")['id'].tolist()
        df_cluster_sim = df1.loc[go_id, go_id].copy()
        df_cluster_sim['std'] = df_cluster_sim.std(axis=1)
        df_cluster_sim.reset_index(drop=False, inplace=True)
        df_cluster_sim.sort_values(by=['std'], inplace=True)
        # 选择标准差最小的三个Go terms
        df_std_top3 = df_cluster_sim.tail(3)[['index']].copy()
        df_std_top3.columns = ['id']
        df_tmp = pd.merge(df_std_top3, df2, how='left', on=['id'])
        df_tmp['cluster'] = [j]*df_tmp.shape[0]
        #print(df_tmp)
        df_return = pd.concat([df_return, df_tmp], axis=0)
        ## 给结点打上std 标签
        df_tmp2 = pd.concat([df_tmp2, df_cluster_sim], axis=0)
    df_tmp2 = df_tmp2[['index', 'std']]
    df_tmp2.columns = ['id', 'std']
    df2 = pd.merge(df2, df_tmp2, how='left', on=['id'])
    df2.sort_values(by=['cluster', 'std'], ascending=(True, False), inplace=True)
    return df2, df_return   


```

```python
## 取数据的平均值
import pandas as pd 
df = pd.read_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq_new/Jin668_glr48_Fov_V991_RNA_Seq_wide.bed", sep="\t")
df[['GeneId', 'Chrom', 'Start', 'End']] = df['indexs'].str.split("-", expand = True)
Times = ['0min', '1min', '5min', '10min', '30min', '90min', '180min', '720min']
df_result = pd.DataFrame()
for t in Times:
    if t == '0min':
        treatmens = ['Glr19_mock', 'J668_mock']
    else:
        treatmens = ['Glr19_mock', 'J668_mock', 'J668_Fov7', 'J668_V991', 'Glr19_Fov7']
    for s in treatmens:
        sample_rep1 = "{0}-{1}-rep1".format(s, t)
        sample_rep2 = "{0}-{1}-rep2".format(s, t)
        sample = "{0}-{1}".format(s, t)
        #print(df[['GeneId', 'Chrom', 'Start', 'End']])
        #print(sample_rep1, sample_rep2)
        df_tmp = df[['GeneId', 'Chrom', 'Start', 'End', sample_rep1, sample_rep2]].copy()
        df_tmp[sample] = df_tmp[[sample_rep1, sample_rep2]].mean(axis=1)
        df_tmp = df_tmp[['Chrom', 'Start', 'End', 'GeneId', sample]].copy()
        if df_result.shape[0] == 0:
            df_result = df_tmp.copy()
        else:
            df_result = pd.merge(df_result, df_tmp, how='inner', on=['Chrom', 'Start', 'End', 'GeneId'])
        #print(df_result.shape[0])
df_result.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq_new/Jin668_glr48_Fov_V991_RNA_Seq_wide_mean.csv", index=False)
            
```

```python
## 分析确定可表达的基因数目
import pandas as pd
# 宽数据转长数据
df = pd.read_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq_new/Jin668_glr48_Fov_V991_RNA_Seq_wide_mean.csv")
df_long = pd.melt(df, id_vars=['GeneId', 'Chrom', 'Start', 'End'])
df_long.columns = ['GeneId', 'Chrom', 'Start', 'End', 'samples', 'TPM']
df_long.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq_new/Jin668_glr48_Fov_V991_RNA_Seq_long_mean.csv", index=False)
# 数据统计，可表达基因数目
df_exp_gene = df_long.query("TPM>=0.5").copy()
df_exp_gene['number'] = [1]*df_exp_gene.shape[0]
df_result = df_exp_gene.groupby(by=['samples'], as_index=False)['number'].sum()
df_result[['sample', 'times']] = df_result['samples'].str.split("-", expand=True)
df_result.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq_new/Jin668_glr48_Fov_V991_expG.csv", index=False)
```

```python
## 统计差异表达基因的数目
import pandas as pd
Control = ['Glr19_mock', 'J668_mock', 'J668_mock', 'J668_Fov7', 'J668_Fov7']
Treatment = ['Glr19_Fov7', 'J668_Fov7', 'J668_V991', 'J668_V991', 'Glr19_Fov7']
df_result = pd.DataFrame(columns=['tags', 'times', 'number'])
for i in range(5):
    c, t = Control[i], Treatment[i]
    for ts in ['1min', '5min', '10min', '30min', '90min', '180min', '720min']:
        input_f = "G:/Billfish/VD_FOV_3Dgenome/RNA_Seq_new/DEG/{0}-{1}-{2}-_DEG.bed".format(c, t, ts)
        tags = "{0}-{1}".format(c, t)
        df = pd.read_csv(input_f, sep="\t")
        df_result.loc[df_result.shape[0], ] = [tags, ts, df.shape[0]]
df_result.to_csv("G:/Billfish/VD_FOV_3Dgenome/RNA_Seq_new/DEG/J668_Glr19_DEG_number.csv", index=False)
    
```
