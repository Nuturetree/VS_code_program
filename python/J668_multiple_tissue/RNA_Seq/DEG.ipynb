```python
## 将亚组间差异表达基因添加表达量
import pandas as pd
import numpy as np
tissues = ['Anther', 'Cotyledon', 'Ovule',  'Hypocotyl', 'Leaf', 'Petal', 'Radicle', 'Root', 'Stem', 'Stigma']
df_hg = pd.read_csv('G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/homo_gene2/Ghir_homo_gene_pairs.txt', sep="\t")
df_hg.columns = ['GeneId_at', 'GeneId_dt']
df_hg_tpm = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/homo_gene2/all_HG_TPM.bed", sep="\t")
df_hg_tpm['At_log2TPM_1'] = df_hg_tpm['At_TPM'].apply(lambda x: np.log2(x+1.5))
df_hg_tpm['Dt_log2TPM_1'] = df_hg_tpm['Dt_TPM'].apply(lambda x: -(np.log2(x+1.5)))
h = 1
def hg_tpm(F_path, df_h, tag1, tag2, df1, df2, tags, sigs):
    df1.columns = [tag1, 'control']
    df2.columns = [tag2, 'treatment']
    df = pd.read_csv(F_path, sep="\t")
    df.rename(columns={'GeneId' : tag1}, inplace=True)
    df = pd.merge(df, df_hg, how='inner', on=[tag1])
    dfA = pd.merge(df, df1, how='inner', on=[tag1])
    df_AA = pd.merge(dfA, df2, how='inner', on=[tag2])
    df_AA['tags'] = [tags]*df_AA.shape[0]
    df_AA['sigs'] = [sigs]*df_AA.shape[0]
    return df_AA
for i in range(9):
    first = tissues[i]
    df_A_At = df_hg_tpm.query("tags==@first")[['GeneId_at', 'At_log2TPM_1']].copy()
    df_A_Dt = df_hg_tpm.query("tags==@first")[['GeneId_dt', 'Dt_log2TPM_1']].copy()
    for j in range(i+1, 10):
        second = tissues[j]
        output_f = "G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/reads/ST_VS_ST/{0}_{1}_subg_TPM.bed".format(first, second)
        df_D_At = df_hg_tpm.query("tags==@second")[['GeneId_at', 'At_log2TPM_1']].copy()
        df_D_Dt = df_hg_tpm.query("tags==@second")[['GeneId_dt', 'Dt_log2TPM_1']].copy()
        ## AA
        input_AA = "G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/reads/ST_VS_ST/AA/{0}_{1}_DEG.bed".format(first, second)
        df_AA = hg_tpm(input_AA, df_hg, 'GeneId_at', 'GeneId_at', df_A_At, df_D_At, 'Same', 'AA')
        ## AD
        input_AD = "G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/reads/ST_VS_ST/AD/{0}_{1}_DEG.bed".format(first, second)
        df_AD = hg_tpm(input_AD, df_hg, "GeneId_at", "GeneId_dt", df_A_At, df_D_Dt, 'Subg', 'AD')
        ## DD
        input_DD = "G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/reads/ST_VS_ST/DD/{0}_{1}_DEG.bed".format(first, second)
        df_DD = hg_tpm(input_DD, df_hg, "GeneId_dt", "GeneId_dt", df_A_Dt, df_D_Dt, 'Same', "DD")
        ## DA
        input_DA = "G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/reads/ST_VS_ST/DA/{0}_{1}_DEG.bed".format(first, second)
        df_DA = hg_tpm(input_DA, df_hg, "GeneId_dt", "GeneId_at", df_A_Dt, df_D_At, 'Subg', "DA")
        df_result = pd.concat([df_AA, df_AD, df_DD, df_DA], axis=0)
        df_result['sig_tags'] = df_result['sig'].map(str) + "-" + df_result['tags'].map(str)
        df_result.to_csv(output_f, sep="\t", index=False)
    
```

```python
### 分析同源基因在组织间差异表达，同时表达模式并不一致，统计两两组织间所有的这种情况
import pandas as pd
## 整理同源基因在组织间差异表达模式
def Hg_tissues_DEG_mode(df_AA, df_DD, df_hg):
    df_AA, df_DD, df_hg = df_AA[['GeneId', 'sig']].copy(), df_DD[['GeneId', 'sig', 'logFC']].copy(), df_hg.copy()
    df_AA.columns = ['GeneId_At', 'sig_At', 'logFC_At']
    df_DD.columns = ['GeneId_Dt', 'sig_Dt', 'logFC_Dt']
    df_A = pd.merge(df_hg, df_AA, how='left', on=['GeneId_At'])
    df_AD = pd.merge(df_A, df_DD, how='left', on=['GeneId_Dt'])
    df_AD.fillna("Not", inplace=True)
    return df_AD

## 整理所有组织间的同源基因差异表达模式
def All_Tissue_DEM():
    tissues = ['anther', 'ovule', 'root',  'radicle', 'leaf', 'hypocotyl', 'cotyledon', 'stem', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'stigma']
    df_hg = pd.read_csv('G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_seq/homo_gene2/Ghir_homo_gene_pairs.txt', sep="\t")
    df_result = pd.DataFrame()
    for i in range(11):
        for j in range(i+1, 12):
            T1 = tissues[i]
            T2 = tissues[j]
            AA_input = "G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue_hg/AA/{0}_{1}_DEG.bed".format(T1, T2)
            DD_input = "G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue_hg/DD/{0}_{1}_DEG.bed".format(T1, T2)
            df_AA = pd.read_csv(AA_input)
            df_DD = pd.read_csv(DD_input)
            df_tmp = Hg_tissues_DEG_mode(df_AA, df_DD, df_hg)
            df_tmp['tissues'] = ["{0}_{1}".format(T1, T2)]*df_tmp.shape[0]
            df_result = pd.concat([df_result, df_tmp], axis=0)
    return df_result
## 同源基因组织差异表达模式
df_Tissues_DEM = All_Tissue_DEM()
df_Tissues_DEM.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue_hg/AA_DD_HDEM/J668_allTissue_DEM.csv", index=False)
```

```python
## 分析在11对组织间差异表达的基因
import pandas as pd
df = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue/statistics/J668_DEG_N_tissues.csv")
## 找到11对能发生差异表达的同源基因
df_11 = df.query("DEG_T_N==11").copy()
df_11.drop(['DEG_T_N', 'number'], axis=1, inplace=True)
df_11.set_index('GeneId', inplace=True)
df_11_T = df_11.T

## 分析组织间差异表达，选最大的比较数目
def sta(df):
    df_index = df.index.to_list()
    tissues1 = [i.split('-')[0] for i in df_index]
    tissues2 = [i.split('-')[1] for i in df_index]
    tissues = tissues1 + tissues2
    df_tissues = pd.DataFrame({'tissues':tissues})
    max_tissues = df_tissues['tissues'].value_counts().idxmax()
    max_tissues_number = df_tissues['tissues'].value_counts().max()
    return max_tissues, max_tissues_number
## 
df_result = pd.DataFrame(columns=['GeneId', 'tissues', 'times'])
for c in df_11_T.columns:
    df_tmp = df_11_T[df_11_T[c]>0][[c]].copy()
    max_T, max_T_N = sta(df_tmp)
    df_result.loc[df_result.shape[0]] = [c, max_T, max_T_N]

df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue/statistics/J668_11_diffTissPair_maxTissue_number.csv", index=False)
```

```python
## 分析在20, 21对组织间差异表达的基因, 
# 存在两种情况，基因A 在组织A中表达量差异高于所有组织，同时在组织B中也高于所有组织，此外其它组织表达水平没明显差异
# 第二：基因A 在组织A中高于除组织B以外所有组织(10),同时组织B也高于所有组织 
import pandas as pd
df = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue/statistics/J668_DEG_N_tissues.csv")
## 找到11对能发生差异表达的同源基因
df_11 = df.query("DEG_T_N==20 | DEG_T_N==21").copy()
df_11.drop(['DEG_T_N', 'number'], axis=1, inplace=True)
df_11.set_index('GeneId', inplace=True)
df_11_T = df_11.T

## 分析组织间差异表达，选最大的比较数目
def sta(df):
    df_index = df.index.to_list()
    tissues1 = [i.split('-')[0] for i in df_index]
    tissues2 = [i.split('-')[1] for i in df_index]
    tissues = tissues1 + tissues2
    df_tissues = pd.DataFrame({'tissues':tissues})
    value_counts = df_tissues['tissues'].value_counts().sort_values(ascending=False)
    max_tissues1 = value_counts.index[0]
    max_tissues_number1 = value_counts.iloc[0]
    max_tissues2 = value_counts.index[1]
    max_tissues_number2 = value_counts.iloc[1]
    return max_tissues1, max_tissues_number1, max_tissues2, max_tissues_number2,
## 
df_result = pd.DataFrame(columns=['GeneId', 'tissues', 'times'])
for c in df_11_T.columns:
    df_tmp = df_11_T[df_11_T[c]>0][[c]].copy()
    max_T1, max_T_N1, max_T2, max_T_N2  = sta(df_tmp)
    df_result.loc[df_result.shape[0]] = [c, max_T1, max_T_N1]
    df_result.loc[df_result.shape[0]] = [c, max_T2, max_T_N2]
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue/statistics/J668_20_21_diffTissPair_maxTissue_number.csv", index=False)
```

```python
## 分析同源基因TPM 在组织中的差异倍数
import pandas as pd
import numpy as np
tissues = ['anther', 'ovule', 'root',  'radicle', 'leaf', 'hypocotyl', 'cotyledon', 'stem', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'stigma']
df_oppsite_gene = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue_hg/AA_DD_HDEM/opposite/J668_HG_tissue_pairs_opposite_complaxheatmap.csv")
df_oppsite_gene = df_oppsite_gene[['GeneId_At']].copy()
df_hg = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/J668_HG_pairs.csv")
df_oppsite_HG = pd.merge(df_hg, df_oppsite_gene, how='inner', on=['GeneId_At'])
print(df_oppsite_HG.shape[0])
df_TPM = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/RNA_Seq_data/J668_all_Tissue_mean_TPM_long.csv")

## 整理同源基因对TPM
def HG_pair_TPM(df_hg, df_TMP, tissue):
    df_hg = df_hg.copy()
    df_TPM = df_TMP[['GeneId', 'TPM']].copy()
    df_TPM.columns = ['GeneId_At', '{}_TPM_At'.format(tissue)]
    df_At = pd.merge(df_hg, df_TPM, how='left', on=['GeneId_At'])
    df_TPM.columns = ['GeneId_Dt', "{}_TPM_Dt".format(tissue)]
    df_At_Dt = pd.merge(df_At, df_TPM, how='left', on=['GeneId_Dt'])
    return df_At_Dt

## 分析同源基因TPM比例
def HG_TPM_ratio(df, T1, T2):
    T1_At, T1_Dt, T2_At, T2_Dt = "{}_TPM_At".format(T1), "{}_TPM_Dt".format(T1), "{}_TPM_At".format(T2), "{}_TPM_Dt".format(T2)
    df['At'] = (df[T1_At]+0.001)/(df[T2_At]+0.001)
    df['Dt'] = (df[T1_Dt]+0.001)/(df[T2_Dt]+0.001)
    df['AA_DD'] = np.log2(df['At']/df['Dt'])
    df_return = df[['GeneId_At', 'GeneId_Dt', 'AA_DD']].copy()
    df_return.columns = ['GeneId_At', 'GeneId_Dt', "{0}_{1}".format(T1,T2)]
    return df_return
## 计算两个组织的TPM 比例
def tissues_TPM_ratio(df_hg):
    df_return = pd.DataFrame()
    for i in range(11):
        for j in range(i+1, 12):
            T1, T2 = tissues[i], tissues[j]
            df_T1 = df_TPM.query("tissues==@T1").copy()
            df_T2 = df_TPM.query("tissues==@T2").copy()
            df_T1_TPM = HG_pair_TPM(df_hg, df_T1, T1)
            df_T2_TPM = HG_pair_TPM(df_hg, df_T2, T2)
            #print(df_T1_TPM)
            df_T1T2_TPM = pd.merge(df_T1_TPM, df_T2_TPM, how='inner', on=['GeneId_At', 'GeneId_Dt'])
            df_tmp = HG_TPM_ratio(df_T1T2_TPM, T1, T2)
            if df_return.shape[0] == 0:
                df_return = df_tmp.copy()
            else:
                df_return = pd.merge(df_return, df_tmp, how='inner', on=['GeneId_At', 'GeneId_Dt'])
    return df_return
df_result = tissues_TPM_ratio(df_oppsite_HG)
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue_hg/AA_DD_HDEM/opposite/J668_HG_tissue_pairs_TPM_opposite_complaxheatmap.csv", index=False)
            
```

```python
## 统计分析差异表达基因，分析AA, DD, AD, DA /Up 和 Down 的占比
import pandas as pd
tissues = ['Anther', 'Cotyledon', 'Ovule',  'Hypocotyl', 'Leaf', 'Petal', 'Radicle', 'Root', 'Stem', 'Stigma']
df_result = pd.DataFrame()
df_hg = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/homo_gene2/Ghir_homo_gene_pairs.txt", sep="\t")
for i in range(9):
    first = tissues[i]
    for j in range(i+1, 10):
        second = tissues[j]
        ab = "{0}_{1}".format(first, second)
        df_tmp = pd.DataFrame()
        for k in ['AA', 'AD', 'DA', 'DD']:
            df_DEG = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/reads/ST_VS_ST/{0}/{1}_DEG.bed".format(k, ab), sep="\t")
            if k =='AA' or k=='AD':
                df_DEG.rename(columns={'GeneId':'GeneId_At'}, inplace=True)
                df_hg_DEG = pd.merge(df_hg, df_DEG, how='inner', on=['GeneId_At'])
            else:
                df_DEG.rename(columns={'GeneId':'GeneId_Dt'}, inplace=True)
                df_hg_DEG = pd.merge(df_hg, df_DEG, how='inner', on=['GeneId_Dt'])
            df_hg_DEG = df_hg_DEG.query("sig!='Not'")[['GeneId_At', 'GeneId_Dt', 'sig']].copy()
            df_hg_DEG.rename(columns={'sig':k}, inplace=True)
            if df_tmp.shape[0] == 0:
                df_tmp = df_hg_DEG
            else:
                df_tmp = pd.merge(df_tmp, df_hg_DEG, how='outer', on=['GeneId_At', 'GeneId_Dt'])
        df_tmp.fillna("Not", inplace=True)
        df_tmp['tags'] = ["{0}_vs_{1}".format(first, second)]*df_tmp.shape[0]
        df_result = pd.concat([df_result, df_tmp], axis=0)
#print(df_result.head())
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/reads/ST_VS_ST/bar_figure/Multi_Subg_DEG.bed", sep="\t", index=False)
```

```python
## 统计各组织，亚基因差异表达基因比例 2
import pandas as pd
df = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/reads/ST_VS_ST/bar_figure/Multi_Subg_DEG.bed", sep="\t")
df_Keep = df.query("(AA=='Up' & AD=='Up' & DA=='Up' & DD=='Up') | (AA=='Down' & AD=='Down' & DA=='Down' & DD=='Down')").copy()
df_Hg = df.query("((AA=='Up' & DD=='Up') & (AD!='Up' | DA!='Up')) | ((AA=='Down' & DD=='Down') & (AD!='Down' | DA!='Down'))").copy()
df_Subg = df.query("((AD=='Up' & DA=='Up') & (AA!='Up' |  DD!='Up')) | ((AD=='Down' & DA=='Down') & (AA!='Down' | DD!='Down'))").copy()
df_Not = df.query("AA=='Not' & AD=='Not' & DA=='Not' & DD=='Not'").copy()

df_dyname = pd.concat([df, df_Keep, df_Hg, df_Subg, df_Not], axis=0)
df_dyname.drop_duplicates(keep=False, inplace=True)
df_Keep['sig'] =['Stable']*df_Keep.shape[0]
df_Hg['sig'] = ['Hg_Stable']*df_Hg.shape[0]
df_Subg['sig'] = ['Subg_Stable']*df_Subg.shape[0]
df_dyname['sig'] = ['Dynamic']*df_dyname.shape[0]
df_result = pd.concat([df_Keep, df_Hg, df_Subg, df_dyname], axis=0)
df_result['number'] = [1]*df_result.shape[0]
df_sta = df_result.groupby(by=['tags', 'sig'], as_index=False)['number'].sum()
df_sta.to_csv('G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/reads/ST_VS_ST/bar_figure/Multi_Subg_DEG_sta2.bed', sep="\t", index=False)
```

```python
## 确定所有组织间差异表达的基因数目
import pandas as pd
tissues = ['anther', 'ovule', 'root', 'radicle', 'leaf', 'hypocotyl', 'cotyledon', 'stem', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'stigma']
# 统计上调和下调的基因数目
def up_down_number(df):
    df_tmp = df.query("sig=='Up' | sig=='Down'").copy()
    df_tmp['number'] = [1]*df_tmp.shape[0]
    df_return = df_tmp.groupby(by=['sig'], as_index=False)['number'].sum()
    return df_return
df_result = pd.DataFrame()
for i in range(11):
    for j in range(i+1, 12):
        input_f = "G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue/{0}_{1}_DEG.csv".format(tissues[i], tissues[j])
        tags = "{0}_{1}".format(tissues[i], tissues[j])
        df = pd.read_csv(input_f)
        df_sta_tmp = up_down_number(df)
        df_sta_tmp['tags'] = [tags]*df_sta_tmp.shape[0]
        df_result = pd.concat([df_result, df_sta_tmp], axis=0)
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue/J668_DEG_number_statistics.csv", index=False)
```

```python
## 分析确定两个亚基因组间比较的差异
import pandas as pd
tissues = ['anther', 'ovule', 'root', 'radicle', 'leaf', 'hypocotyl', 'cotyledon', 'stem', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'stigma']

def statis_number(tags):
    df_return = pd.DataFrame(columns = ['number', 'tags', 'tissues'])
    for i in range(11):
        for j in range(i+1, 12):
            first = tissues[i]
            last = tissues[j]
            input_f = "G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue_hg/{0}/{1}_{2}_DEG.bed".format(tags, first, last)
            df_tmp = pd.read_csv(input_f)
            Up_number = df_tmp.query("sig=='Up'").shape[0]
            df_return.loc[df_return.shape[0], ] = [Up_number, tags, "{0}_{1}".format(first, last)]
    return df_return
df_AA_up = statis_number("AA")
df_DD_up = statis_number("DD")
df_result = pd.concat([df_AA_up, df_DD_up], axis=0)
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue_hg/J668_AA_DD_up_number.csv", index=False)
```
