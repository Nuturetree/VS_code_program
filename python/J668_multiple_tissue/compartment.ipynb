```python
# 将CALDER整体compartment 切分为windows bin的文件格式
import pandas as pd
# the result of CALDER convert to single windows
def single_bin(df, sample, winds):
    df_return = pd.DataFrame('chr', 'start', 'end', 'tags')
    for index, row in df.iterrows():
        c, s, e, t = row['chr'], row['start'], row['end'], row['tags']
        for i in range(s, e, winds):
            s1, e1 = s-1, s-1+winds
            df_return.loc[df_return.shape[0]] = [c, s1, e1, t]
    df_return.columns = ['chr', 'start', 'end', sample]
    return df_return

# multiple sample analysis
samples = ['anthers', 'cotyledon', 'fiber_5DPA', 'fiber_10DPA', 'fiber_20DPA', 'hypocotyl', 'leaf', 'petals', 'radicle', 'root', 'stem', 'stigma']
```

```python
# cooltools indentify compartment
# import standard python libraries
import numpy as np
import pandas as pd
import os, subprocess
import sys
# Import python package for working with cooler files and tools for analysis
import cooler
import cooltools.lib.plotting
import cooltools
import bioframe

input_f = sys.argv[1]
gene_f = sys.argv[2]
output_f = sys.argv[3]
clr_input = "{}::resolution/100000".format(input_f)
clr = cooler.Cooler(clr_input)
gc_cov = pd.read_csv(gene_f, sep="\t")
view_df = pd.DataFrame({'chrom': clr.chromnames, 'start': 0, 'end': clr.chromsizes.values, 'name': clr.chromnames})
# obtain first 3 eigenvectors
cis_eigs = cooltools.eigs_cis( clr, gc_cov, view_df=view_df, n_eigs=3,)
# cis_eigs[0] returns eigenvalues, here we focus on eigenvectors
eigenvector_track = cis_eigs[1][['chrom','start','end','E1']]
# write compartment
eigenvector_track.to_csv(output_f, sep='\t', index=False)
```
