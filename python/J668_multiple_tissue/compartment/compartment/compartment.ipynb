```python
## 统计分析两个亚基因组，stable A， stable B 和dynamic 区间所占的比例
import pandas as pd
df = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/sta_dynamic_cpt/Dynamic_stable_compartment_number.csv")
def stable_AB_dynamic(df, sg):
    total_number = df['bin_number'].sum()
    stable_A = df.query("tags==0")['bin_number'].sum()
    stable_B = df.query("tags==12")['bin_number'].sum()
    print(stable_A, stable_B)
    dynamic = df.query("tags!=0 & tags!=12")['bin_number'].sum()
    stable_A_ratio = (stable_A/total_number)*100
    stable_B_ratio = (stable_B/total_number)*100
    dymamic_ratio = (dynamic/total_number)*100
    df_return = pd.DataFrame({"ratio":[stable_A_ratio, stable_B_ratio, dymamic_ratio], "tags":['stable_A', 'stable_B', 'dynamic'], "subg":[sg, sg, sg]})    
    return df_return
df_At = df.query("subg=='At'").copy() 
df_Dt = df.query("subg=='Dt'").copy()
df_return_At = stable_AB_dynamic(df_At, 'At')
df_return_Dt = stable_AB_dynamic(df_Dt, 'Dt')
df_result = pd.concat([df_return_At, df_return_Dt], axis=0)
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/sta_dynamic_cpt/Dynamic_stable_compartment_ratio.csv", index=False)
 
```

```python
## A-A, A-B, B-B 随机抽取两千来分析
import pandas as pd
df_result = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/WGCNA/J668_WGCNA_cpt.csv")
df_result['number'] = [1]*df_result.shape[0]
df_t = df_result.groupby(by=['tags'], as_index=False)['number'].sum()
print(df_t)
df_AA = df_result.query("tags=='A_A'").copy()
df_AB = df_result.query("tags=='A_B'").copy()
df_BB = df_result.query("tags=='B_B'").copy()

def sample_average(df, n, tags):
    df = df[['weight']].copy()
    df_return = pd.DataFrame()
    for i in range(n):
        df_tmp = df.sample(3000)
        df_tmp.reset_index(drop=True, inplace=True)
        #print(df_tmp.head())
        if df_return.shape[0]==0:
            df_return = df_tmp.copy()
        else:
            df_return = pd.concat([df_return , df_tmp], ignore_index=True, axis=1)
    df_return['weight'] = df_return.mean(axis=1)
    df_return['tags'] = [tags]*df_return.shape[0]
    return df_return[['tags', 'weight']]
df_AA_s = sample_average(df_AA, 100, 'A_A')
df_AB_s = sample_average(df_AB, 100, 'A_B')
df_BB_s = sample_average(df_BB, 100, 'B_B')
print(df_AA_s.shape[0])
df = pd.concat([df_AA_s, df_AB_s, df_BB_s], axis=0)
df.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/WGCNA/J668_WGCNA_cpt_sample.csv", index=False)
```

```python
## 分析同源基因A/B compartment 状态
import pandas as pd
df = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/WGCNA/J668_all_tissue_gene_cpt_status.csv", names=['chrom', 'start', 'end', 'GeneId', 'tissue', 'cpt'])
tissues = ['anther', 'leaf', 'petal', 'radicle', 'root', 'stem', 'stigma', 'fiber_10DPA', 'fiber_20DPA', 'hypocotyl', 'cotyledon', 'ovule']
def Hg_cpt_status(df):
    df_hg = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/homo_gene2/Ghir_homo_gene_pairs.txt", sep="\t")
    df_a = df[['GeneId', 'tissue', 'cpt']].copy()
    df_a.drop_duplicates(keep='first', subset=['GeneId'], inplace=True)
    df_a.columns = ['GeneId_At', 'tissue', 'cpt_At']    
    df_tmp1 = pd.merge(df_hg, df_a, how='left', on=['GeneId_At'])
    df_b = df_a[['GeneId_At', 'cpt_At']].copy()
    df_b.columns = ['GeneId_Dt', 'cpt_Dt']
    df_return = pd.merge(df_tmp1, df_b, how='left', on=['GeneId_Dt'])
    df_return.dropna(inplace=True)
    return df_return
df_result = pd.DataFrame()
for t in tissues:
    df_tissue = df.query("tissue==@t").copy()
    df_tmp = Hg_cpt_status(df_tissue)
    df_result = pd.concat([df_result, df_tmp], axis=0)
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/subg/J668_HG_cpt_status.csv",  index=False)
```

```python
## 分析有多少亚基因组bias基因存在compartment差异
import pandas as pd
# 1、各同源基因的A/B compartment 状态
def HG_cpt_statu(df_cpt, df_hg):
    df_cpt, df_hg = df_cpt.copy(), df_hg.copy()
    # At genome
    df_cpt = df_cpt[['GeneId', 'tissues', 'cpt']].copy()
    df_cpt.columns = ['GeneId_At', 'tissues', 'cpt_At'].copy()
    df_a = pd.merge(df_cpt, df_hg, how='inner', on=['GeneId_At'])
    #df_cpt = df_cpt[['GeneId_At', 'cpt_At']].copy()
    df_cpt.columns = ['GeneId_Dt', 'tissues', 'cpt_Dt']
    df_return = pd.merge(df_a, df_cpt, how='inner', on=['GeneId_Dt', 'tissues'])
    return df_return  
# 2、确定同源基因间的bias关系
def HG_bias_statu(df_hg, df_bias):
    df_hg, df_bias = df_hg.copy(), df_bias.copy()
    #print(df_bias)
    df_return = pd.merge(df_hg, df_bias, how='inner', on=['GeneId_At', 'GeneId_Dt', 'tissues'])
    return df_return
    #print(df_return)
# 3、统计各组织与cpt状态差异相关的同源基因bias
def cpt_bias_statu(df):
    df = df[['cpt_At', 'cpt_Dt', 'sig', 'tissues']].copy()
    df_cpt = df.query("cpt_At != cpt_Dt").copy()
    df_cpt['number'] = [1]*df_cpt.shape[0]
    #df['number'] = [1]*df.shape[0]
    df_statu = df_cpt.groupby(by=['tissues'], axis=False)['number'].sum()
    return df_cpt, df_statu
    
df_cpt = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/WGCNA/J668_all_tissue_gene_cpt_status.csv", names=['chrom', 'start', 'end', 'GeneId', 'tissues', 'cpt'])
df_hg = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/homo_gene2/Ghir_homo_gene_pairs.txt", sep="\t")
df_bias = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/homo_gene2/gene_bias/all_tissue_bias_gene_pair.bed", sep="\t")
df_HG_cpt= HG_cpt_statu(df_cpt, df_hg)
df_HG_cpt_bias = HG_bias_statu(df_HG_cpt,df_bias)
df_HG, df_HG_statu = cpt_bias_statu(df_HG_cpt_bias)
df_HG.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/subg/J668_diff_cpt_bias_gene.bed", sep="\t", index=False)
df_HG_statu.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/subg/J668_diff_cpt_bias_gene_statu.bed", sep="\t")
```
