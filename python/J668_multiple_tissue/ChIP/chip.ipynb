```python
# diffbind 附近(左右各500 bp)所包含的基因表达水平上的差异
import numpy as np
import pandas as pd
def diffpeaks_TAU(df_diff_peaks, df_TAU, region):
    df_diff_peaks['start'] = df_diff_peaks['start'] - region
    df_diff_peaks['end'] = df_diff_peaks['end'] + region
    df_return = pd.DataFrame()
    for index, row in df_diff_peaks.iterrows():
        df_tmp = pd.DataFrame()
        C, S, E, T = row['chrs'], row['start'], row['end'], row['log2FC']
        df_tmp = df_TAU.query("chrs==@C and start>=@S and end<=@E").copy()
        if df_tmp.shape[0] > 0:
            if T >0:
                df_tmp['sigs'] = ['Hypocotyl']*df_tmp.shape[0]
            else:
                df_tmp['sigs'] = ['Cotyledon']*df_tmp.shape[0]
            df_return = pd.concat([df_return, df_tmp], axis=0)
    df_return.columns = ['index', 'treatment','value', 'chrs', 'start', 'end', 'class']
    return df_return[['value', 'class', 'treatment', 'index']]

df_diff_peaks = pd.read_csv('G:/Billfish/J668_multip_tissue_3D-genome/Chip_Seq/diffbind/hypocotyl_cotyledon/result_edgeR_sig.bed', sep="\t", names=['chrs', 'start', 'end', 'a', 'log2FC'])
df_TAU = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/RNA_seq/expression_gene/J668_all_tissue_average_TPM_long.txt", sep="\t")
df_TAU_tissue = df_TAU.query("tags=='cotyledon'|tags=='hypocotyl'").copy()
df_result = diffpeaks_TAU(df_diff_peaks, df_TAU_tissue, 5000)
df_result['value'] = np.log2(df_result['value']+0.001)
df_result['treatment'].replace(['hypocotyl', 'cotyledon'], ['Hypocotyl', "Cotyledon"], inplace=True)
print(df_result.shape[0])
df_result.drop_duplicates(keep='first', inplace=True)
print(df_result.shape[0])
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/Chip_Seq/diffbind/hypocotyl_cotyledon/diff_peaks_TAU_10k.bed", sep="\t", index=False)
# print(df_result)
```

```python
## 分析有多少组织差异表达基因在promoter区域存在差异, 统计有多少基因符合差异表达与ChIP-Seq peaks 信号存在与否有关
import pandas as pd
tissues = ['ovule', 'root', 'radicle', 'hypocotyl', 'cotyledon', 'stem', 'stigma']

#df 表示差异表达基因，df_T1表示组织1在promoter区域有信号注释的基因，df_T2表示组织2在promoter区域有信号注释的基因
def add_chip_signal(df, df_T1, df_T2):
    df, df_T1, df_T2 = df.copy(), df_T1.copy(), df_T2.copy()
    df_a = pd.merge(df, df_T1, how='left', on=['GeneId'])
    df_ab = pd.merge(df_a, df_T2, how='left', on=['GeneId'])
    return df_ab
## 分析有peaks注释促进基因上调表达，没有peaks注释导致基因下调表达
def DEG_peaks(df):
    df = df.copy()
    df.fillna(0, inplace=True)
    df_all = df.query("sig!='Not' & control != treatment")
    df_result = df.query("(sig=='Down' & control==1 & treatment==0) | (sig=='Up' & control==0 & treatment==1)").copy()
    print(df_all.shape[0], df_result.shape[0])
    return df_result, df_all

df_H3K4me3 = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/Chip_Seq/peaks/annotation/J668_H3K27ac_promoter_peaks.csv")
df_result = pd.DataFrame()
df_result2 = pd.DataFrame(columns=['control', 'treatment', 'all', 'positive']) ##记录chip peaks 的差异所能够影响的所有差异表达基因，其中包含了正向调控和负向调控
for i in range(6):
    for j in range(i+1, 7):
        T1,T2 = tissues[i], tissues[j]
        input_f = 'G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue/{0}_{1}_DEG.csv'.format(T1,T2)
        df = pd.read_csv(input_f)
        df_T1 = df_H3K4me3.query("tissues ==@T1")[['GeneId']].drop_duplicates().copy()
        df_T2 = df_H3K4me3.query("tissues==@T2")[['GeneId']].drop_duplicates().copy()
        df_T1['control'] = [1]*df_T1.shape[0]
        df_T2['treatment'] = [1]*df_T2.shape[0]
        df_tmp = add_chip_signal(df, df_T1, df_T2)
        #print(T1, T2)
        df_tmp, df_tmp2  = DEG_peaks(df_tmp)
        df_result2.loc[df_result2.shape[0], ] = [T1, T2, df_tmp2.shape[0],  df_tmp.shape[0]]   
        df_tmp['C_tissue'] = [T1]*df_tmp.shape[0]
        df_tmp['T_tissue'] = [T2]*df_tmp.shape[0]
        df_result = pd.concat([df_result, df_tmp], axis=0)
df_result2.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue/epigenome/J668_DEG_H3K27ac_sta.csv", index=False)
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/DEG/tissue/epigenome/J668_DEG_H3K27ac.csv", index=False)
```
