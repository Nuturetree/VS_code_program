```python
# A 列
##1、 整理每个材料A/D亚组， A/B compartment 比例，subcompartment, 三类TAD比例， 组织高表达TAD数目，三类loop数目，cis元件数目，活性和非活性元件数目，表达和不表达基因数目，组织特异高表达基因数目
import pandas as pd

# # 中心的圆比例
# def center():
#     df_return = pd.DataFrame(columns=['tissues', 'subg', 'tags', 'sigs', 'level', 'ratio'])
#     tissues = ['anther', 'ovule', 'leaf', 'stigma', 'stem', 'hypocotyl', 'cotyledon', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'root', 'radicle']
#     for t in tissues:
#         for s in ['a', 'b']:
#             for i in ['At', 'Dt']:
#                 df_return.loc[df_return.shape[0], ] = [t, i, 'center', s, '"""1"""', 0.5]
#     return df_return
def compartment():
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/sta_dynamic_cpt/J668_compartment_statistics_subg.csv"
    df = pd.read_csv(input_f)
    #print(df.head())
    tissues = ['anther', 'ovule', 'leaf', 'stigma', 'stem', 'hypocotyl', 'cotyledon', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'root', 'radicle']
    df_result = pd.DataFrame()
    for t in tissues:
        for s in ['At', 'Dt']:
            df_T = df.query("tissue == @t & subg == @s").copy()
            number_total = df_T['number'].sum()
            df_T['ratio'] = df_T['number']/number_total
            df_T['tags'] = ['compartment']*df_T.shape[0]
            df_T['level'] = ['"""1"""']*df_T.shape[0]
            df_result = pd.concat([df_result, df_T], axis=0)
    df_return = df_result[['tissue', 'subg', 'tags', 'cpt', 'level', 'ratio']].copy()
    df_return.columns = ['tissues', 'subg', 'tags', 'sigs', 'level', 'ratio']
    return df_return
def subcompartment():
    tissues = ['anther', 'ovule', 'leaf', 'stigma', 'stem', 'hypocotyl', 'cotyledon', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'root', 'radicle']
    input_f = 'G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/CALDER2/statistics/J668_tissues_subcpt_sta_subg.bed'
    df = pd.read_csv(input_f, sep="\t")
    tissues = ['anther', 'ovule', 'leaf', 'stigma', 'stem', 'hypocotyl', 'cotyledon', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'root', 'radicle']
    df_result = pd.DataFrame()
    for t in tissues:
        for s in ['At', 'Dt']:
            df_T = df.query("tissues == @t & subg == @s").copy()
            number_total = df_T['number'].sum()
            df_T['ratio'] = df_T['number']/number_total
            df_T['tags'] = ['subcompartment']*df_T.shape[0]
            df_T['level'] = ['"""2"""']*df_T.shape[0]
            df_result = pd.concat([df_result, df_T], axis=0)
    #print(df_result)
    df_return = df_result[['tissues', 'subg', 'tags', 'sigs', 'level', 'ratio']].copy()
    return df_return
## 分析活性TAD的占比
def active_TAD():
    tissues = ['anther', 'ovule', 'leaf', 'stigma', 'stem', 'hypocotyl', 'cotyledon', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'root', 'radicle']
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/TAD_new/TSG_TAD/J668_TSG_TAD_length_sta.csv"
    df = pd.read_csv(input_f)
    df_result = pd.DataFrame()
    for t in tissues:
        for s in ['At', 'Dt']:
            df_T = df.query("tissues == @t & subg == @s").copy()
            number_total = df_T['number'].sum()
            df_T['ratio'] = df_T['number']/number_total
            df_T['tags'] = ['TSG_TAD']*df_T.shape[0]
            df_T['level'] = ['"""3"""']*df_T.shape[0]
            df_result = pd.concat([df_result, df_T], axis=0)
    df_result = df_result[['tissues', 'subg', 'tags', 'sigs', 'level', 'ratio']].copy()
    return df_result
## 各组织三类TAD基因组长度占比
def TAD():
    tissues = ['anther', 'ovule', 'leaf', 'stigma', 'stem', 'hypocotyl', 'cotyledon', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'root', 'radicle']
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/TAD_new/TAD_type/J668_TAD_type_length.csv"
    df = pd.read_csv(input_f)
    df_result = pd.DataFrame()
    for t in tissues:
        for s in ['At', 'Dt']:
            df_T = df.query("tissues == @t & subg == @s").copy()
            number_total = df_T['number'].sum()
            df_T['ratio'] = df_T['number']/number_total
            df_T['tags'] = ['TAD_type']*df_T.shape[0]
            df_T['level'] = ['"""4"""']*df_T.shape[0]
            df_result = pd.concat([df_result, df_T], axis=0)
    df_result = df_result[['tissues', 'subg', 'tags', 'sigs', 'level', 'ratio']].copy()
    return df_result
## 分析三种不同的loop类型在各组织中的占比
def loop():
    tissues = ['anther', 'ovule', 'leaf', 'stigma', 'stem', 'hypocotyl', 'cotyledon', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'root', 'radicle']
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/loops/loop_type/J668_loop_type_statis_subg.csv"
    df = pd.read_csv(input_f)
    df_result = pd.DataFrame()
    for t in tissues:
        for s in ['At', 'Dt']:
            df_T = df.query("tissues == @t & subg == @s").copy()
            number_total = df_T['number'].sum()
            df_T['ratio'] = df_T['number']/number_total
            df_T['tags'] = ['TAD_type']*df_T.shape[0]
            df_T['level'] = ['"""5"""']*df_T.shape[0]
            df_result = pd.concat([df_result, df_T], axis=0)
    df_result = df_result[['tissues', 'subg', 'tags', 'sigs', 'level', 'ratio']].copy()
    return df_result
## 分析活性cis bin
def cis():
    tissues = ['anther', 'ovule', 'leaf', 'stigma', 'stem', 'hypocotyl', 'cotyledon', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'root', 'radicle']
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/loops/Cis-element/J668_active_cis_sta.csv"
    df = pd.read_csv(input_f)
    df_result = pd.DataFrame()
    for t in tissues:
        for s in ['At', 'Dt']:
            df_T = df.query("tissues == @t & subg == @s").copy()
            number_total = df_T['number'].sum()
            df_T['ratio'] = df_T['number']/number_total
            df_T['tags'] = ['Cis']*df_T.shape[0]
            df_T['level'] = ['"""6"""']*df_T.shape[0]
            df_result = pd.concat([df_result, df_T], axis=0)
    df_result = df_result[['tissues', 'subg', 'tags', 'sigs', 'level', 'ratio']].copy()
    return df_result
## 分析表达和不表达同源基因
def hg():
    tissues = ['anther', 'ovule', 'leaf', 'stigma', 'stem', 'hypocotyl', 'cotyledon', 'fiber_10DPA', 'fiber_20DPA', 'petal', 'root', 'radicle']
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/new_RNA_Seq/homo_gene2/J668_hg_exp_nexp_statistics.csv"
    df = pd.read_csv(input_f)
    df_result = pd.DataFrame()
    for t in tissues:
        for s in ['At', 'Dt']:
            df_T = df.query("tissues == @t & subg == @s").copy()
            number_total = df_T['number'].sum()
            df_T['ratio'] = df_T['number']/number_total
            df_T['tags'] = ['HG']*df_T.shape[0]
            df_T['level'] = ['"""7"""']*df_T.shape[0]
            df_result = pd.concat([df_result, df_T], axis=0)
    df_result = df_result[['tissues', 'subg', 'tags', 'sigs', 'level', 'ratio']].copy()
    return df_result

#df_center = center()
df_compartment = compartment()
df_subcpt = subcompartment()
df_active_TAD = active_TAD()
df_TAD = TAD()
df_loop = loop() 
df_cis = cis()
df_hg = hg()
df_result = pd.concat([df_compartment, df_subcpt, df_active_TAD, df_TAD, df_loop, df_cis, df_hg], axis=0)
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/model/J668_model.csv", index=False)
```

```python
# B列
import pandas as pd
## 分析保守和动态的结构
def conserved_dynamic(df, tags):
    df = df.copy()
    if tags == 'compartment':
        chroms = set(df['chrom'].to_list())
        for c in chroms:
            
            df_conserved = df[(df=='A').all(axis=1) | (df=='B').all(axis=1)]

# 分析保守compartment和动态compartment
def compartment():
    input_f = 'G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/sta_dynamic_cpt/J668_compartment_wide.csv'
    df = pd.read_csv(input_f)
    df_result = pd.DataFrame(columns=['subg', 'chrom', 'tags', 'number'])
    for i in ['Ghir_A', 'Ghir_D']:
        df_subg = df.query("chrom.str.contains(@i)", engine='python').copy()
        chroms = set(df_subg['chrom'].to_list())
        for c in chroms:
            df_chr = df_subg.query("chrom==@c").copy()
            df_chr.drop(['chrom', 'start', 'end'], axis=1, inplace=True)
            df_conserved = df_chr[((df_chr=='A').all(axis=1)) | ((df_chr=='B').all(axis=1))]
            df_dynamic = df_chr[~(((df_chr=='A').all(axis=1)) | ((df_chr=='B').all(axis=1)))]
            df_result.loc[df_result.shape[0], ] = [i, c, 'conserved', df_conserved.shape[0]]
            df_result.loc[df_result.shape[0], ] = [i, c, 'dynamic', df_dynamic.shape[0]]
    df_result['sigs'] = ['compartment']*df_result.shape[0]
    return df_result
## 分析sub compartment 占比
def subcompartment():
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/CALDER2/subcompartment/J668_allTissues_subcompartment_wide.bed"
    df = pd.read_csv(input_f, sep='\t')
    df_result = pd.DataFrame(columns=['subg', 'chrom', 'tags', 'number'])
    for i in ['Ghir_A', 'Ghir_D']:
        df_subg = df.query("chrom.str.contains(@i)", engine='python').copy()
        chroms = set(df_subg['chrom'].to_list())
        for c in chroms:
            df_chr = df_subg.query("chrom==@c").copy()
            df_chr.drop(['chrom', 'start', 'end'], axis=1, inplace=True)
            df_conserved = df_chr[((df_chr=='A1').all(axis=1)) | ((df_chr=='A2').all(axis=1)) | ((df_chr=='A3').all(axis=1)) | ((df_chr=='A4').all(axis=1)) | ((df_chr=='B1').all(axis=1)) | ((df_chr=='B2').all(axis=1)) | ((df_chr=='B3').all(axis=1)) | ((df_chr=='B4').all(axis=1))]
            df_dynamic = df_chr[~(((df_chr=='A1').all(axis=1)) | ((df_chr=='A2').all(axis=1)) | ((df_chr=='A3').all(axis=1)) | ((df_chr=='A4').all(axis=1)) | ((df_chr=='B1').all(axis=1)) | ((df_chr=='B2').all(axis=1)) | ((df_chr=='B3').all(axis=1)) | ((df_chr=='B4').all(axis=1)))]
            df_result.loc[df_result.shape[0], ] = [i, c, 'conserved', df_conserved.shape[0]]
            df_result.loc[df_result.shape[0], ] = [i, c, 'dynamic', df_dynamic.shape[0]]
    df_result['sigs'] = ['subcompartment']*df_result.shape[0]    
    return df_result
# 分析保守和动态TAD 边界的占比
def TAD():
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/TAD_new/conserved_TAD/J668_Mutiple_tissues_boundary_conserved_statistics.csv"
    df = pd.read_csv(input_f)
    df['subg'] = df['chrom'].apply(lambda x: 'Ghir_A' if 'Ghir_A' in x else 'Ghir_D')
    df['tags2'].replace(['special'], ['dynamic'], inplace=True)
    df_return = pd.DataFrame(df[['subg', 'chrom', 'tags2']].value_counts(), columns=['number'])
    df_return.reset_index(inplace=True)
    df_return['sigs'] = ['boundary']*df_return.shape[0]
    df_return.rename(columns={'tags2':'tags'}, inplace=True)
    return df_return
# 分析与基因相关的保守loop和特异性loops
def loops():
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/loops/loop_type/J668_loop_type.bed"
    df = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/HiC/loops/loop_type/J668_loop_type.bed", sep="\t")
    df_G = df.query("tags.str.contains('G')", engine='python')[['chrom1', 'start1', 'end1', 'chrom2', 'start2', 'end2']].copy()
    df_G_sta = pd.DataFrame(df_G[['chrom1', 'start1', 'end1', 'chrom2', 'start2', 'end2']].value_counts(), columns=['number'])
    df_G_sta.reset_index(inplace=True)
    df_G_sta['tags'] = df_G_sta['number'].apply(lambda x: 'conserved' if x > 1 else 'dynamic')
    df_G_sta['subg'] = df_G_sta['chrom1'].apply(lambda x: 'Ghir_A' if 'Ghir_A' in x else "Ghir_D")
    df_return = df_G_sta.groupby(by=['subg', 'chrom1', 'tags'], as_index=False)['number'].sum()
    df_return['sigs'] = ['loop']*df_return.shape[0]
    df_return.rename(columns={'chrom1':'chrom'}, inplace=True)
    return df_return
# # 分析组织cis调控元件
def cis():
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/loops/Cis-element/J668_active_cis.csv"
    df = pd.read_csv(input_f)
    df_active = df.query("sigs=='active'", engine='python')
    df_sta = pd.DataFrame(df_active[['chrom', 'start', 'end']].value_counts(), columns=['number'])
    df_sta.reset_index(inplace=True)
    df_sta['tags'] = df_sta['number'].apply(lambda x: 'conserved' if x > 1 else 'dynamic')
    df_sta['subg'] = df_sta['chrom'].apply(lambda x: "Ghir_A" if "Ghir_A" in x else "Ghir_D")
    df_return = df_sta.groupby(by=['subg', 'chrom', 'tags'], as_index=False)['number'].sum()
    df_return['sigs'] = ["Cis"]*df_return.shape[0]
    return df_return
    
df_compartment = compartment()
df_subcompartment = subcompartment()
df_TAD = TAD()
df_loop = loops()
df_cis = cis()
df_result = pd.concat([df_compartment, df_subcompartment, df_TAD, df_loop, df_cis], axis=0)
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/model/J668_module_B.csv", index=False)
```

```python
## 分析保守的和动态的compartment，subcompartment和TAD分别占基因组大小
# 1、 保守TAD
import pandas as pd
def TAD():
    At_length = 1413909078
    Dt_length = 820082591
    total_At_length = 12*1413909078
    total_Dt_length = 12*820082591
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/TAD_new/conserved_TAD/J668_conserved_TAD.csv"
    df = pd.read_csv(input_f)
    df_conserved = df.query("tags=='conserved'", engine='python')[['chr', 's', 'e']].copy()
    df_conserved['subg'] = df_conserved['chr'].apply(lambda x: 'At' if 'Ghir_A' in x else 'Dt')
    df_conserved['length'] = df_conserved['e'] - df_conserved['s']
    df_sta = df_conserved.groupby(by=['subg'], as_index=False)['length'].sum()
    df_sta.rename(columns={'length' : 'conserved_TAD'}, inplace=True)
    df_sta['total_length'] = [total_At_length, total_Dt_length]
    df_sta['chrom_length'] = [At_length, Dt_length]
    df_sta['dynamic_TAD'] = df_sta['total_length'] - df_sta['conserved_TAD']
    df_sta['conserved_ratio'] = df_sta['conserved_TAD']/df_sta['total_length']
    df_sta['dynamic_ratio'] = df_sta['dynamic_TAD']/df_sta['total_length']
    df_sta['conserved'] = df_sta['conserved_ratio']*df_sta['chrom_length']
    df_sta['dynamic'] = df_sta['dynamic_ratio']*df_sta['chrom_length']
    df_tmp = df_sta[['subg', 'conserved', 'dynamic']].copy()
    df_result = df_tmp.melt(id_vars=['subg'])
    df_result.columns = ['subg', 'sigs', 'length']
    df_result['tags'] = ['TAD']*df_result.shape[0]
    return df_result
def compartment():
    input_f = 'G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/sta_dynamic_cpt/J668_compartment_wide.csv'
    df = pd.read_csv(input_f)
    At_length = 1413909078
    Dt_length = 820082591
    df_result = pd.DataFrame(columns=['subg', 'sigs', 'length'])
    for i in ['Ghir_A', 'Ghir_D']:
        df_subg = df.query("chrom.str.contains(@i)", engine='python').copy()
        df_subg.drop(['chrom', 'start', 'end'], axis=1, inplace=True)
        df_conserved = df_subg[((df_subg=='A').all(axis=1)) | ((df_subg=='B').all(axis=1))]
        conserved_length = df_conserved.shape[0]*100000
        if i == 'Ghir_A':
            dynamic_length = At_length - conserved_length
        else:
            dynamic_length = Dt_length - conserved_length
        df_result.loc[df_result.shape[0],] = [i, 'conserved', conserved_length]
        df_result.loc[df_result.shape[0],] = [i, 'dynamic', dynamic_length]
    df_result['subg'].replace(['Ghir_A', 'Ghir_D'], ['At', 'Dt'], inplace=True)
    df_result['tags'] = ['compartment']*df_result.shape[0]
    return df_result
def subcompartment():
    input_f = 'G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/CALDER2/subcompartment/J668_allTissues_subcompartment_wide.bed'
    df = pd.read_csv(input_f, sep="\t")
    At_length = 1413909078
    Dt_length = 820082591
    df_result = pd.DataFrame(columns=['subg', 'sigs', 'length'])
    for i in ['Ghir_A', 'Ghir_D']:
        df_subg = df.query("chrom.str.contains(@i)", engine='python').copy()
        df_subg.drop(['chrom', 'start', 'end'], axis=1, inplace=True)
        df_conserved = df_subg[((df_subg=='A1').all(axis=1)) | ((df_subg=='A2').all(axis=1)) | ((df_subg=='A3').all(axis=1)) | ((df_subg=='A4').all(axis=1)) | ((df_subg=='B1').all(axis=1)) | ((df_subg=='B2').all(axis=1)) | ((df_subg=='B3').all(axis=1)) | ((df_subg=='B4').all(axis=1))]
        conserved_length = df_conserved.shape[0]*10000
        if i == 'Ghir_A':
            dynamic_length = At_length - conserved_length
        else:
            dynamic_length = Dt_length - conserved_length
        df_result.loc[df_result.shape[0],] = [i, 'conserved', conserved_length]
        df_result.loc[df_result.shape[0],] = [i, 'dynamic', dynamic_length]
    df_result['subg'].replace(['Ghir_A', 'Ghir_D'], ['At', 'Dt'], inplace=True)
    df_result['tags'] = ['subcompartment']*df_result.shape[0]
    print(df_result)
    return df_result

df_TAD = TAD()
df_compartment = compartment()
df_subcompartment = subcompartment()
df_result = pd.concat([df_TAD, df_compartment, df_subcompartment], axis=0)
df_result.to_csv('G:/Billfish/J668_multip_tissue_3D-genome/model/J668_module3.csv', index=False)
```

```python
## 分析组织共有loops、cis和组织特有loop和cis
import pandas as pd

# 中心的圆比例
def center():
    df_return = pd.DataFrame(columns=['subg', 'tags', 'sigs', 'level', 'ratio'])
    for s in ['conserved', 'dynamic']:
        df_return.loc[df_return.shape[0], ] = ['a', 'center', s, '"""1"""', 0.5]
    # print(df_return)
    return df_return
# 分析组织共有和组织特有的loop 数目
def loops():
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/loops/loop_type/J668_loop_type.bed"
    df = pd.read_csv("G:/Billfish/J668_multip_tissue_3D-genome/HiC/loops/loop_type/J668_loop_type.bed", sep="\t")
    #df_G = df.query("tags.str.contains('G')", engine='python')[['chrom1', 'start1', 'end1', 'chrom2', 'start2', 'end2']].copy()
    df_G_sta = pd.DataFrame(df[['chrom1', 'start1', 'end1', 'chrom2', 'start2', 'end2']].value_counts(), columns=['number'])
    df_G_sta.reset_index(inplace=True)
    df_G_sta['tags'] = df_G_sta['number'].apply(lambda x: 'conserved' if x > 1 else 'dynamic')
    df_G_sta['subg'] = df_G_sta['chrom1'].apply(lambda x: 'At' if 'Ghir_A' in x else "Dt")
    df_G_sta = df_G_sta.groupby(by=['subg', 'tags'], as_index=False)['number'].sum()
    At_loop_number = df_G_sta.query("subg=='At'")['number'].sum()
    Dt_loop_number = df_G_sta.query("subg=='Dt'")['number'].sum()
    print(At_loop_number, Dt_loop_number)
    df_G_sta['total_number'] = [At_loop_number, At_loop_number, Dt_loop_number, Dt_loop_number]
    df_G_sta['ratio'] = df_G_sta['number']/df_G_sta['total_number']
    df_G_sta['level'] = ['"""2"""', '"""2"""', '"""3"""', '"""3"""']
    df_G_sta = df_G_sta[['subg', 'tags', 'ratio', 'level']].copy()
    df_G_sta['sigs'] = ['loops']*df_G_sta.shape[0]
    df_G_sta.rename(columns={'tags':'sigs', 'sigs':'tags'}, inplace=True)
    return df_G_sta
# 分析组织共有和特有的cis调控元件
def cis():
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/loops/Cis-element/J668_active_cis.csv"
    df = pd.read_csv(input_f)
    df_active = df.query("sigs=='active'", engine='python')
    df_sta = pd.DataFrame(df_active[['chrom', 'start', 'end']].value_counts(), columns=['number'])
    df_sta.reset_index(inplace=True)
    #print(df_sta)
    df_sta['tags'] = df_sta['number'].apply(lambda x: 'conserved' if x > 1 else 'dynamic')
    df_sta['subg'] = df_sta['chrom'].apply(lambda x: "At" if "Ghir_A" in x else "Dt")
    df_sta = df_sta.groupby(by=['subg', 'tags'], as_index=False)['number'].sum()
    At_cis_number = df_sta.query("subg=='At'")['number'].sum()
    Dt_cis_number = df_sta.query("subg=='Dt'")['number'].sum()
    df_sta['total_number'] = [At_cis_number, At_cis_number, Dt_cis_number, Dt_cis_number]
    df_sta['ratio'] = df_sta['number']/df_sta['total_number']
    df_sta['level'] = ['"""4"""', '"""4"""', '"""5"""', '"""5"""']
    df_sta = df_sta[['subg', 'tags', 'ratio', 'level']].copy()
    df_sta['sigs'] = ['cis']*df_sta.shape[0]
    df_sta.rename(columns={'tags':'sigs', 'sigs':'tags'}, inplace=True)
    return df_sta
# 分析
df_center = center()
df_loops = loops()
df_cis = cis()
df_result = pd.concat([df_center, df_loops, df_cis], axis=0)
df_result.to_csv('G:/Billfish/J668_multip_tissue_3D-genome/model/J668_module4.csv', index=False)
```

```python
## 分析动态compartment、subcompartment、TAD、loop和cis
import pandas as pd
def compartment():
    input_f = 'G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/cworld/100K/sta_dynamic_cpt/J668_compartment_wide.csv'
    df = pd.read_csv(input_f)
    T = ['anther', 'cotyledon', 'fiber_10DPA', 'fiber_20DPA', 'hypocotyl', 'leaf', 'ovule', 'petal', 'radicle', 'root', 'stem', 'stigma']
    df.replace(['A', 'B'], [1, 0], inplace=True)
    df['total'] = df[T].sum(axis=1)
    df = df.query("0<total<12").copy()
    df['subg'] = df['chrom'].apply(lambda x: 'At' if 'Ghir_A' in x else 'Dt')
    df_return = pd.DataFrame(df[['subg', 'total']].value_counts(), columns=['number'])
    df_return.reset_index(inplace=True)
    df_return['tags'] = ['compartment']*df_return.shape[0]
    df_return.rename(columns={'total':'sigs'}, inplace=True)
    df_return['number'] = df_return['number']*10
    print(df_return)
    return df_return
def subcompartment():
    input_f = 'G:/Billfish/J668_multip_tissue_3D-genome/HiC/compartment/CALDER2/subcompartment/J668_allTissues_subcompartment_wide.bed'
    df = pd.read_csv(input_f, sep="\t")
    df_return = pd.DataFrame()
    for c in ['Ghir_A', 'Ghir_D']:
        df_subg = df.query("chrom.str.contains(@c)").copy()
        for i in ['A1', 'A2', 'A3', 'A4', 'B1', 'B2', 'B3', 'B4']:
            df_sta = pd.DataFrame((df_subg==i).sum(axis=1), columns=['tags'])
            df_sta = pd.DataFrame(df_sta['tags'].value_counts())
            df_sta.reset_index(inplace=True)
            df_sta.columns = ['sigs', 'number']
            df_sta = df_sta.query("0<sigs<12").copy()
            df_sta['tags'] = [i]*df_sta.shape[0]
            df_sta['subg'] = [c]*df_sta.shape[0]
            df_return = pd.concat([df_return, df_sta], axis=0)
    df_return['subg'].replace(['Ghir_A','Ghir_D'], ['At', 'Dt'], inplace=True)
    print(df_return)
    return df_return

df_cpt = compartment()
df_subcpt = subcompartment()
df_result = pd.concat([df_cpt, df_subcpt], axis=0)
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/model/J668_module6.csv", index=False)
```

```python
## 分析组织共有boundary、cis、loop
import pandas as pd
def boundary():
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/TAD_new/conserved_TAD/J668_conserved_boundary_all_statistics.csv"
    df = pd.read_csv(input_f, names=['chrom', 'boundary', 'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'total'])
    df['subg'] = df['chrom'].apply(lambda x: 'At' if 'Ghir_A' in x else 'Dt')
    df_sta = pd.DataFrame(df[['subg', 'total']].value_counts(), columns=['number'])
    df_sta['tags'] = ['boundary']*df_sta.shape[0]
    df_sta.reset_index(inplace=True)
    df_sta['total'] = df_sta['total'] + 1
    return df_sta
def loop():
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/loops/loop_type/J668_loop_type_all.csv"
    df = pd.read_csv(input_f)
    df.replace(['N-G'], ['G-N'], inplace=True)
    df_sta = pd.DataFrame(df[['chrom1', 'start1', 'end1', 'chrom2', 'start2', 'end2', 'loop_type']].value_counts(), columns=['total'])
    df_sta.reset_index(inplace=True)
    df_sta['subg'] = df_sta['chrom1'].apply(lambda x: 'At' if 'Ghir_A' in x else 'Dt')
    df_return = pd.DataFrame(df_sta[['subg','loop_type', 'total']].value_counts(), columns=['number'])
    df_return.reset_index(inplace=True)
    df_return.rename(columns={'loop_type':'tags'}, inplace=True)
    return df_return
def cis():
    input_f = "G:/Billfish/J668_multip_tissue_3D-genome/HiC/loops/Cis-element/J668_active_cis.csv"
    df = pd.read_csv(input_f)
    df_active = df.query("sigs=='active'").copy()
    df_sta = pd.DataFrame(df_active[['subg', 'chrom', 'start', 'end']].value_counts(), columns=['total'])
    df_sta.reset_index(inplace=True)
    df_return = pd.DataFrame(df_sta[['subg', 'total']].value_counts(), columns=['number'])
    df_return.reset_index(inplace=True)
    df_return['tags'] = ['cis']*df_return.shape[0]
    return df_return

df_loop = loop()
df_boundary = boundary()
df_cis = cis()
df_result = pd.concat([df_boundary, df_loop, df_cis], axis=0)
df_result = df_result.query("total>1").copy()
df_result.to_csv("G:/Billfish/J668_multip_tissue_3D-genome/model/J668_module7.csv", index=False)
```
